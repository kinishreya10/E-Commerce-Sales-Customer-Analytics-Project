```python
#Look at the statuses that appear: If most are canceled, created, or unavailable
#Then the missing approval date is expected.Action: Keep the nulls and document the reason.

#If many are delivered or shipped - That's unusual because an order generally needs approval before
# moving forward. Then inspect those rows:
orders[
    (orders["order_status"].isin(["delivered"])) &
    (orders["order_approved_at"].isnull())
]
```

```python
# Remove 14 anomalous records od delivered
orders = orders[
    ~(
        (orders["order_status"] == "delivered") &
        (orders["order_approved_at"].isnull())
    )
]
```

```python
#now check again if its removed or not
orders.loc[
    orders["order_approved_at"].isnull(),
    "order_status"
].value_counts()

# Missing Approval Dates
#Orders with status "canceled" and "created" were allowed to retain null approval timestamps because approval may 
#never have occurred. 14 delivered orders contained missing approval timestamps, which is inconsistent with the 
#expected business process. These records were removed as data-quality anomalies.
```

```python
#2.4 --- Convert Date columns (Datetime Conversion)
orders.dtypes
```

```python
# Convert 
date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])
```

```python
# Now verify if they're convert or not
orders.dtypes
```

```python
#After converting dates, create these columns:
# 2.5 Feature engineering
#2.5 --- Create Analysis Feature
orders["purchase_year"] = (
    orders["order_purchase_timestamp"].dt.year
)

orders["purchase_month"] = (
    orders["order_purchase_timestamp"].dt.month
)

orders["purchase_day"] = (
    orders["order_purchase_timestamp"].dt.day
)
```

```python
# Now check
orders[
    ["order_purchase_timestamp",
     "purchase_year",
     "purchase_month",
     "purchase_day"]
].head()
```

```python
# 2.6 --- Create Delivery Metrics (v.imp)- These become KPIs later.
# Delivery Time
orders["delivery_days"] = (
    orders["order_delivered_customer_date"]
    -
    orders["order_purchase_timestamp"]
).dt.days
```

```python
# Delivery Delay
orders["delivery_delay_days"] = (
    orders["order_delivered_customer_date"]
    -
    orders["order_estimated_delivery_date"]
).dt.days
```

```python
# Delayed Order Flag
orders["is_delayed"] = np.where(
    orders["delivery_delay_days"] > 0,
    1,
    0
)
```

```python
# Now check- 
orders["is_delayed"].value_counts()
#o/p- ex- [-4 = Delivered 4 days early],[0 = On time],[3 = Delivered 3 days later]
```

```python
# 2.7 Quick Validation
orders[
    [
        "delivery_days",
        "delivery_delay_days"
    ]
].describe()
```

```python
#What the no.s mean  - delivery_days
#o/p- avg- -12 days, median- 10 days, max- 209 days

#delivery_delay_days
#Mean- -11.88, median- -12, max- 188 now interpretation:Negative values mean:
#Actual Delivery Date   <   Estimated Delivery Date

#In simple words, Most orders were delivered earlier than promised. Ex- 
#Estimated: 20 Jan
#Delivered: 10 Jan
#Delay = -10
#That's why your average delay is negative.
#This is actually a good business insight.
```
